```python
###########
# Functions

@typecheck
def main(filename: str, log: List[Genre]) -> None:
    """
    Reads the file from given filename, analyzes the data, and returns a scatterplot
    of the average rating of webtoon genres compared to the popularity
    (defined as average number of subscribers in thousands for stories in the genre) 
    """
    # Template from HtDAP, based on function composition 
    return show_scatterplot(read(filename), log) 

@typecheck
def show_scatterplot(lowd: List[WebtoonData], log: List[Genre]) -> None:
    """
    creates a scatterplot of the average rating of stories in a
    genre vs. genre popularity (defined as average number of subscribers
    in thousands for stories in the genre).
    
    Assumes that lengths of average ratings, popularity, and counts are the same
    """
    # return None # stub
    # template based on visualization
    
    popularity = av_subs_all_genres(lowd, log)
    av_rating = av_rating_all_genres(lowd, log)
    counts = list_of_counts(lowd, log)
    
    # set the labels for the axes
    plt.xlabel('Popularity (Average Number of Subscribers in 1000s)')
    plt.ylabel('Average Rating')
    plt.title('Graph of Genre Popularity and Average Rating of Webtoons')
    
    # range for the axes
    # [x-min, x-max, y-min, y-max]
    plt.axis([0,1000.0,8.5,10.0])
    
    plt.scatter(popularity,av_rating,marker='o', c='palevioletred', s=counts)
    plt.show()     

    
@typecheck
def read(filename: str) -> List[WebtoonData]:
    """    
    reads information from the specified file and returns a list of Webtoon Data (includes
    the Genre of each story, the number of subscribers it has in thousands, and the rating it has out of 10).
    """
    # return []  #stub
    # Template from HtDAP
    # lowd contains the result so far
    lowd = [] # type: List[WebtoonData]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            wd = WebtoonData(parse_genre(row[4]), parse_subs(row[6]), parse_float(row[5]))
            lowd.append(wd)
    
    return lowd


@typecheck
def parse_genre(s: str) -> Genre:
    """
    Given a string, returns its equivalent Genre. Must be one of Romance, Supernatural, Fantasy, Action, 
    Thriller, Mystery, Comedy, Heartwarming, Slice of life, Sci-fi,
    Historical, Sports, Superhero, Informative, Horror, or Drama.
    """
    # return Genre.Romance # stub
    # return...(s) # template
    
    if s == "Romance":
        return Genre.Romance
    elif s == "Supernatural":
        return Genre.Supernatural
    elif s == "Fantasy":
        return Genre.Fantasy
    elif s == "Action":
        return Genre.Action
    elif s == "Thriller":
        return Genre.Thriller
    elif s == "Mystery":
        return Genre.Mystery
    elif s == "Comedy":
        return Genre.Comedy
    elif s == "Heartwarming":
        return Genre.Heartwarming
    elif s == "Slice of life":
        return Genre.Slice_of_life
    elif s == "Sci-fi":
        return Genre.Sci_fi
    elif s == "Historical":
        return Genre.Historical
    elif s == "Sports":
        return Genre.Sports
    elif s == "Superhero":
        return Genre.Superhero
    elif s == "Informative":
        return Genre.Informative
    elif s == "Horror":
        return Genre.Horror
    elif s == "Drama":
        return Genre.Drama

@typecheck
def parse_subs(subs: str) -> float:
    """
    Given a string of numbers representing number of subscribers, returns number of subscribers as float in thousands.
    If string ends with "M", meaning the value is currently in millions, removes "M", parses into float, and multiplies
    value by 1000 to convert into thousands.
    If string ends with "K", meaning value is currently in thousands, removes "K" and parses into float.
    """
    # return 0.0 # stub
    # template based on atomic non-distinct
    if subs[-1] == "M":
        return parse_float(subs[:-1]) * 1000
    elif subs[-1] == "K":
        return parse_float(subs[:-1])
       
    
@typecheck
def list_of_counts(lowd: List[WebtoonData], log: List[Genre]) -> List[int]:
    """
    given a list of webtoon data and a list of Genres, returns a list of the number of webtoons in each genre
    
    ASSUME: log contains all unique genres found in lowd
    ASSUME: log does NOT have any genres that are not present in lowd
    """
    # return [] # stub
    # Template based on List[Genre] with additional parameter and reference rule
    
    # loc represents the number of webtoons seen in the Genre so far
    loc = [] # type: List[int]
    for g in log:
        loc = loc + [count_in_genre(lowd, g)]
    
    return loc    
    
    
@typecheck
def av_rating_all_genres(lowd: List[WebtoonData], log: List[Genre]) -> List[float]:
    """
    given a list of webtoon data and a list of genres, returns a list of the average rating for webtoons in each
    genre in the list
    ASSUME: log contains all unique genres present in lowd
    ASSUME: log does NOT contain genres that are not in lowd; log only has genres found in lowd
    """
    # return [] # stub
    # template based on List[Genre] with additional parameter and reference rule
    
    # loa represents the list of average ratings so far
    loa = [] # type: List[float]
    for g in log:
        loa = loa + [average_rating(lowd, g)]
        
    return loa    
    
@typecheck
def av_subs_all_genres(lowd: List[WebtoonData], log: List[Genre]) -> List[float]:
    """
    given a list of webtoon data and a list of genres, returns a list of the average number of subscribers in thousands
    for webtoons in each genre in the list
    ASSUME: log contains all unique genres present in lowd
    ASSUME: log does NOT contain genres that are not in lowd; log only has genres found in lowd
    """
    # return [] # stub
    # template based on List[Genre] with additional parameter and reference rule
    
    # loa represents the list of average ratings so far
    loa = [] # type: List[float]
    for g in log:
        loa = loa + [average_subs(lowd, g)]
        
    return loa
    
    
@typecheck
def average_rating(lowd: List[WebtoonData], g: Genre) -> float:
    """
    given a list of WebtoonData and a Genre, returns the average rating of all Webtoons in the Genre
    ASSUME: The list of WebtoonData has at least 1 item with the same Genre as the given Genre
    """
    # return 0.0 # stub
    # template based on composition and reference rule
    # Plan:
    # 1. filter the list by Genre
    # 2. count all the items in the filtered list
    # 3. find sum of all the ratings in the filtered list
    # 4. calculate the average rating by dividing the sum of ratings by the number of webtoons in the genre

    filtered_list = filter_by_genre(lowd, g)
    total_in_genre = count_webtoons(filtered_list)
    sum_of_ratings = sum_ratings(filtered_list)
    av_rating = sum_of_ratings/total_in_genre
    
    return av_rating    
    

@typecheck
def average_subs(lowd: List[WebtoonData], g: Genre) -> float:
    """
    given a list of WebtoonData, returns the average number of subscribers for webtoons in a given genre
    ASSUME: the list of webtoon data has at least 1 item that has the same genre as the given genre
    """
    # return 0.0 # stub
    # template based on composition and reference rule
    # Plan: 
    # 1. filter list by genre
    # 2. count number of items in filtered list
    # 3. find the total number of subscribers to webtoons in the genre
    # 4. calculate the average number of subscribers by dividing the total number of subscribers by the number of items
    
    filtered_list = filter_by_genre(lowd, g)
    total_webtoons = count_webtoons(filtered_list)
    total_subs = sum_subs(filtered_list)
    average_subs = total_subs/total_webtoons
    return average_subs
    
    
@typecheck    
def sum_ratings(lowd: List[WebtoonData]) -> float:
    """
    given a list of WebtoonData, returns the sum of all the ratings of the WebtoonData
    """
    # return 0.0 # stub
    # template based on List[WebtoonData] and reference rule
    
    # sum_ratings represents the sum of all ratings seen so far
    sum_ratings = 0.0 # type: float
    
    for wd in lowd:
        sum_ratings = add_ratings(sum_ratings, wd)
    
    return sum_ratings
    
    
@typecheck
def add_ratings(tot: float, wd2: WebtoonData) -> float:
    """
    given a WebtoonData and a float, returns the sum
    """
    # return 0.0 # stub
    # template based on WebtoonData with additional parameter
    
    return tot + wd2.rating


@typecheck
def sum_subs(lowd: List[WebtoonData]) -> float:
    """
    given a list of WebtoonData, returns the sum of the subscribers in the list
    """
    # return 0.0 # stub
    # template based on List[WebtoonData]
    
    # subs represents the sum of the subscribers of the webtoon data seen so far
    subs = 0.0 # type: float
    
    for wd in lowd:
        subs = add_subscribers(subs, wd)
    
    return subs
    
@typecheck
def add_subscribers(subs: float, wd: WebtoonData) -> float:
    """
    given a float and a WebtoonData, returns the sum of the subscribers in thousands
    """
    # return 0.0 # stub
    # template based on WebtoonData with additional parameter
    
    return subs + wd.subs    

@typecheck
def count_webtoons(lowd: List[WebtoonData]) -> int:
    """
    given a list of WebtoonData, counts the number of WebtoonData in the list.
    """
    # return 0 # stub
    # template based on List[WebtoonData]
    
    # tot represents all the webtoons seen so far
    
    tot = 0 # type: int
    
    for wd in lowd:
        tot = tot + 1
    
    return tot
      
    
@typecheck
def filter_by_genre(lowd: List[WebtoonData], g: Genre) -> List[WebtoonData]:
    """
    given a list of webtoon data and a Genre, filters and returns a list of webtoon data of the same genre.
    """
    # return [] # stub
    # template based on List[WebtoonData] with additional parameter and reference rule
    
    # filtered is all the webtoon data seen so far with the correct genre
    
    filtered = [] # Type: List[WebtoonData]
    
    for wd in lowd:
        if same_genre(wd, g) == True:
            filtered = filtered + [wd]
    
    return filtered
        
@typecheck
def count_in_genre(lowd: List[WebtoonData], g: Genre) -> int:
    """
    returns the number of webtoons in the list that have the same genre as a given genre
    """
    # return 0 # stub
    # template based on List[WebtoonData] with additional paramter and reference rule
    
    # count represents the total so far
    count = 0 # type: int
    
    for wd in lowd:
        if same_genre(wd, g):
            count = count + 1
    
    return count
    
    
@typecheck
def same_genre(wd: WebtoonData, g: Genre) -> bool:
    """
    returns True if genre of Webtoon matches given genre and False otherwise
    """
    # return True # stub
    # template based on WebtoonData with additional parameter and reference rule
    
    return is_genre(wd.genre, g)    
    
    
@typecheck
def is_genre(g1: Genre, g2: Genre) -> bool:
    """
    returns True if a Genre (g1) is the same as a given genre (g2) and False otherwise
    """
    # return True # stub
    # template based on Genre (2 times)
    if g1 == Genre.Romance:
        return g2 == Genre.Romance
    elif g1 == Genre.Supernatural:
        return g2 == Genre.Supernatural
    elif g1 ==  Genre.Fantasy:
        return g2 == Genre.Fantasy
    elif g1 == Genre.Action:
        return g2 == Genre.Action
    elif g1 == Genre.Thriller:
        return g2 == Genre.Thriller
    elif g1 == Genre.Mystery:
        return g2 == Genre.Mystery
    elif g1 == Genre.Comedy:
        return g2 == Genre.Comedy
    elif g1 == Genre.Heartwarming:
        return g2 == Genre.Heartwarming
    elif g1 == Genre.Slice_of_life:
        return g2 == Genre.Slice_of_life
    elif g1 == Genre.Sci_fi:
        return g2 == Genre.Sci_fi
    elif g1 == Genre.Historical:
        return g2 == Genre.Historical
    elif g1 == Genre.Sports:
        return g2 == Genre.Sports
    elif g1 == Genre.Superhero:
        return g2 == Genre.Superhero
    elif g1 == Genre.Informative:
        return g2 == Genre.Informative
    elif g1 == Genre.Horror:
        return g2 == Genre.Horror
    elif g1 == Genre.Drama:
        return g2 == Genre.Drama    
    

# tests for main function
# NOTE: WILL NOT ADJUST SCALE/AXES/AREA OF DOTS FOR TESTS -> ACTUAL DATASET WILL BE MUCH LARGER AND MORE VISIBLE
## IMPORTANT: DUE TO SMALL SAMPLE SIZE, DOTS FOR SOME GENRES WILL NOT FIT INTO AXES OF PLOT, WHICH HAS BEEN
## ADJUSTED TO VISIBLY SHOW RESULTS FOR THE LARGE DATASET. TESTING WAS USED TO SEE IF FUNCTION WAS DESIGNED CORRECTLY.

start_testing()
expect(main("TestingShort.csv", [Genre.Romance, Genre.Supernatural]), None)
expect(main("TestingShort2.csv", [Genre.Fantasy, Genre.Supernatural, Genre.Action,
                                  Genre.Romance, Genre.Drama]), None)
# for test 2, using TestingShort2.csv, you would expect to see 5 small dots that take up an area equal to 1 webtoon.
# these dots would be positioned at (x, y): (3000, 9.82),(649, 9.66), (537.6, 9.87),(1100, 9.82),(4300, 9.78)
# because of the constraint of the adjusted axes, only dots positioned at (649, 9.66) and (537.6, 9.87) are visible in
# the tests.
 
summary()    
    
# tests for read
start_testing()

expect(read("TestingEmpty.csv"), [])
expect(read("TestingShort.csv"), [WebtoonData(Genre.Romance, 4200.0, 9.62), 
                                  WebtoonData(Genre.Romance, 6400.0, 9.6),
                                  WebtoonData(Genre.Romance, 2100.0, 9.81),
                                  WebtoonData(Genre.Romance, 3500.0, 9.79),
                                  WebtoonData(Genre.Supernatural, 1500.0, 9.85)])

expect(read("TestingShort2.csv"), [WebtoonData(Genre.Fantasy, 3000.0, 9.82),
                                   WebtoonData(Genre.Supernatural, 649.0, 9.66),
                                   WebtoonData(Genre.Action, 537.6, 9.87),
                                   WebtoonData(Genre.Romance, 1100.0, 9.82),
                                   WebtoonData(Genre.Drama, 4300.0, 9.78)])

# testing summary for read
summary()    
    

# tests for show scatterplot
start_testing()

expect(show_scatterplot([], []),None)
expect(show_scatterplot(LOWD4, [Genre.Romance, Genre.Thriller]), None)    

# testing summary
summary()
    
    
    
# Testing for other functions (helpers for main/show_scatterplot/read)
start_testing()


# tests for parse subs
expect(parse_subs("1.5M"), 1500.0)
expect(parse_subs("1M"), 1000.0)
expect(parse_subs("796.4K"), 796.4)
expect(parse_subs("8.1K"), 8.1)


# tests for parse genre
expect(parse_genre("Romance"), Genre.Romance) 
expect(parse_genre("Supernatural"), Genre.Supernatural) 
expect(parse_genre("Fantasy"), Genre.Fantasy) 
expect(parse_genre("Action"), Genre.Action)
expect(parse_genre("Thriller"), Genre.Thriller) 
expect(parse_genre("Mystery"), Genre.Mystery) 
expect(parse_genre("Comedy"), Genre.Comedy) 
expect(parse_genre("Heartwarming"), Genre.Heartwarming) 
expect(parse_genre("Slice of life"), Genre.Slice_of_life) 
expect(parse_genre("Sci-fi"), Genre.Sci_fi) 
expect(parse_genre("Historical"), Genre.Historical) 
expect(parse_genre("Sports"), Genre.Sports) 
expect(parse_genre("Superhero"), Genre.Superhero) 
expect(parse_genre("Informative"), Genre.Informative) 
expect(parse_genre("Horror"), Genre.Horror)
expect(parse_genre("Drama"), Genre.Drama)


# tests for list of counts
expect(list_of_counts([], []), [])
expect(list_of_counts(LOWD1, [Genre.Romance, Genre.Thriller]), [1,1])
expect(list_of_counts([WD1, WD2, WebtoonData(Genre.Romance, 367.5, 9.0)], [Genre.Romance, Genre.Thriller]), [2,1])

# tests for getting average rating of all genres
expect(av_rating_all_genres([], []), [])
expect(av_rating_all_genres([WD1, WD2], LOG1), [9.0])
expect(av_rating_all_genres([WD1, WebtoonData(Genre.Supernatural, 10.6, 9.8)], LOG2), [9.0, 9.8])
expect(av_rating_all_genres([WD1, WebtoonData(Genre.Romance, 10.6, 9.8)], [Genre.Romance]), [9.4])

# tests for average subscribers for all genres
expect(av_subs_all_genres([], []), [])
expect(av_subs_all_genres([WD1, WD2], LOG1), [367.5])
expect(av_subs_all_genres([WD1, WebtoonData(Genre.Supernatural, 10.6, 9.8)], LOG2), [367.5, 10.6])
expect(av_subs_all_genres([WD1, WebtoonData(Genre.Romance, 10.6, 9.8)], [Genre.Romance]), [189.05])

# test for getting the average rating for a given genre
expect(average_rating(LOWD3, Genre.Fantasy), 7.8/1)
expect(average_rating([WD1, WD2, WD4, WebtoonData(Genre.Romance, 367.5, 10.0),
                            WebtoonData(Genre.Romance, 5050.1, 6.0)], Genre.Romance), 25.0/3)


# test for average subs
expect(average_subs(LOWD3, Genre.Fantasy), 4800.0/1)
expect(average_subs([WD1, WD2, WD4, WebtoonData(Genre.Romance, 0.0, 10.0),
                            WebtoonData(Genre.Romance, 16.5, 6.0)], Genre.Romance), (367.5 + 0.0 + 16.5)/3)


# test for returning the sum of ratings in a list of webtoon data
expect(sum_ratings([]), 0.0)
expect(sum_ratings(LOWD1), 9.0 + 9.8)
expect(sum_ratings([WD3, WD4]), 7.8 + 8.4)
expect(sum_ratings(LOWD3), 9.0 + 9.8 + 7.8 + 8.4)


# test for adding ratings together
expect(add_ratings(9.0, WD2), 9.0 + 9.8)
expect(add_ratings(7.8, WD2), 7.8 + 9.8)
expect(add_ratings(8.4, WebtoonData(Genre.Horror, 17.2, 8.4)), 8.4 + 8.4)


# test for sum of subscribers
expect(sum_subs([]), 0.0)
expect(sum_subs(LOWD1), 367.5 + 10.6)
expect(sum_subs([WD3, WD4]), 4800.0 + 17.2)
expect(sum_subs(LOWD3), 367.5 + 10.6 + 4800.0 + 17.2)


# test for adding subscribers
expect(add_subscribers(1000.0, WD1), 1000.0 + 367.5)
expect(add_subscribers(0.0, WD2), 10.6)
expect(add_subscribers(1000.0, WebtoonData(Genre.Horror, 17.2, 8.4)), 1000.0 + 17.2)


# test for counting number of items in list

expect(count_webtoons([]), 0)
expect(count_webtoons(LOWD1), 2)
expect(count_webtoons(LOWD3), 4)
expect(count_webtoons([WebtoonData(Genre.Romance, 367.5, 9.0), 
                       WebtoonData(Genre.Thriller, 10.6, 9.8),
                       WebtoonData(Genre.Fantasy, 4800.0, 7.8)]), 3)

# test for filtering by genre

expect(filter_by_genre([], Genre.Fantasy), [])
expect(filter_by_genre(LOWD1, Genre.Fantasy), [])
expect(filter_by_genre(LOWD1, Genre.Romance), [WD1])
expect(filter_by_genre([WebtoonData(Genre.Romance, 367.5, 9.0), 
                        WebtoonData(Genre.Thriller, 10.6, 9.8),
                        WebtoonData(Genre.Fantasy, 4800.0, 7.8),
                        WebtoonData(Genre.Horror, 17.2, 8.4)], Genre.Romance), 
                       [WebtoonData(Genre.Romance, 367.5, 9.0)])

# tests for counting items in a genre
expect(count_in_genre([], Genre.Romance), 0)
expect(count_in_genre([WD1, WD2, WD3], Genre.Thriller), 1)
expect(count_in_genre([WD1, WebtoonData(Genre.Romance, 367.5, 9.0)], Genre.Romance), 2)

# tests for same genre

expect(same_genre(WD1, Genre.Romance), True)
expect(same_genre(WD2, Genre.Romance), False)
expect(same_genre(WD3, Genre.Fantasy), True)
expect(same_genre(WebtoonData(Genre.Horror, 17.2, 8.4), Genre.Fantasy), False)

# tests for is genre

expect(is_genre(Genre.Romance, Genre.Romance), True)
expect(is_genre(Genre.Supernatural, Genre.Supernatural), True)
expect(is_genre(Genre.Fantasy, Genre.Romance), False)
expect(is_genre(Genre.Action, Genre.Romance), False)
expect(is_genre(Genre.Thriller, Genre.Thriller), True)
expect(is_genre(Genre.Mystery, Genre.Thriller), False)
expect(is_genre(Genre.Comedy, Genre.Comedy), True)
expect(is_genre(Genre.Heartwarming, Genre.Comedy), False)
expect(is_genre(Genre.Slice_of_life, Genre.Slice_of_life), True)
expect(is_genre(Genre.Sci_fi, Genre.Romance), False)
expect(is_genre(Genre.Historical, Genre.Romance), False)
expect(is_genre(Genre.Sports, Genre.Sports), True)
expect(is_genre(Genre.Superhero, Genre.Informative), False)
expect(is_genre(Genre.Informative, Genre.Informative), True)
expect(is_genre(Genre.Horror, Genre.Horror), True)
expect(is_genre(Genre.Drama, Genre.Drama), True)

summary()


```
